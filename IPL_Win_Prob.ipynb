```python
# drop columns having target_runs as NaN
match = match.dropna(subset=['target_runs'])
```

```python
# -----------------------
# 1) Merge match info onto deliveries
# -----------------------
cols_to_merge = ['id','city','team1','team2','winner','target_runs','target_overs']
merged = delivery.merge(match[cols_to_merge], left_on='match_id', right_on='id', how='left')
```

```python
merged.head()
```

```python
# -----------------------
# 2) Keep only 2nd innings (chase) snapshots — common for win-prob models
# -----------------------
merged = merged[merged['inning'] == 2].copy()
```

```python
merged.columns
```

```python
# Cumulative runs scored so far
merged['runs_so_far'] = merged.groupby('match_id')['total_runs'].cumsum()
```

```python
# Cumulative wickets fallen so far
merged['wickets_fallen'] = merged.groupby(['match_id'])['is_wicket'].cumsum()
```

```python
merged.head()
```

```python
# Compute runs_left, balls_left, wickets

# Total balls in innings
merged['total_balls'] = merged['target_overs'] * 6

# Balls bowled till now
merged['balls_bowled'] = (merged['over'] * 6 + merged['ball']).astype(int)

# Remaining runs and balls
merged['runs_left'] = (merged['target_runs'] - merged['runs_so_far']).clip(lower=0)
merged['balls_left'] = (merged['total_balls'] - merged['balls_bowled']).clip(lower=0)

# Remaining wickets
merged['wickets'] = (10 - merged['wickets_fallen']).clip(lower=0)
```

```python
# Compute CRR(current run rate) & RRR(required run rate)

# Current Run Rate
merged['crr'] = np.where(
    merged['balls_bowled'] > 0,
    merged['runs_so_far'] * 6 / merged['balls_bowled'],
    0
)

# Required Run Rate
merged['rrr'] = np.where(
    merged['balls_left'] > 0,
    merged['runs_left'] * 6 / merged['balls_left'],
    0
)
```

```python
# result label: 1 if batting team eventually won the match, else 0
merged['result'] = np.where(
    merged['winner'] == merged['batting_team'],
    1,
    0
)
```

```python
final_df = merged[['batting_team','bowling_team','city',
                   'runs_left','balls_left','wickets',
                   'target_runs','crr','rrr','result']].copy()

```

```python
final_df.isna().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
# Unify team names
replace_dict = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
    'Kings XI Punjab': 'Punjab Kings'
}

for col in ['batting_team', 'bowling_team']:
    final_df[col] = final_df[col].replace(replace_dict)
```

```python
final_df.head()
```

```python
# shuffle
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
X = final_df.iloc[:, :-1]
y = final_df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.describe()
```

```python
X_train
```

```python
# OHE
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
# import models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score, log_loss, confusion_matrix, classification_report
```

```python
# Helper function to evaluate models
def evaluate_model(name, model, X_test, y_test):
    preds = model.predict(X_test)
    probs = model.predict_proba(X_test)[:,1]
    print(f"\n🔹 {name} Results:")
    print(f"Accuracy: {accuracy_score(y_test, preds):.4f}")
    print(f"ROC-AUC:  {roc_auc_score(y_test, probs):.4f}")
    print(f"Log Loss: {log_loss(y_test, probs):.4f}")
    print("\nConfusion Matrix:\n", confusion_matrix(y_test, preds))
    print("\nClassification Report:\n", classification_report(y_test, preds))
    return {
        'Model': name,
        'Accuracy': accuracy_score(y_test, preds),
        'ROC_AUC': roc_auc_score(y_test, probs),
        'LogLoss': log_loss(y_test, probs)
    }

results = []
```

```python
def match_summary(row):
    print("Batting Team-" + row['batting_team'] + " | Bowling Team-" + row['bowling_team'] + " | Target- " + str(row['target_runs']))
```

```python
def match_progression(x_df,id,pipe):
    match = x_df[x_df['id'] == id]
    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wickets','target_runs','crr','rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)

    target = temp_df['target_runs'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]

    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
```

```python
# ==========================================
# 🔹 Model 1: Logistic Regression + GridSearchCV
# ==========================================
```

```python
log_reg = Pipeline([
    ('preprocess', trf),
    ('model', LogisticRegression(solver='liblinear', max_iter=1000))
])
```

```python
param_grid_lr = {
    'model__C': [0.01, 0.1, 1, 10],
    'model__penalty': ['l1','l2']
}
```

```python
grid_lr = GridSearchCV(
    log_reg, param_grid_lr, scoring='roc_auc',
    cv=5, n_jobs=-1, verbose=1
)
```

```python
grid_lr.fit(X_train, y_train)

print("✅ Best Logistic Regression Params:", grid_lr.best_params_)
results.append(evaluate_model("Logistic Regression", grid_lr.best_estimator_, X_test, y_test))
```

```python
# ==========================================
# 🔹 Model 2: Random Forest + GridSearchCV
# ==========================================

rf = Pipeline([
    ('preprocess', trf),
    ('model', RandomForestClassifier(random_state=42))
])

param_grid_rf = {
    'model__n_estimators': [100, 200],
    'model__max_depth': [6, 10, 15],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}

grid_rf = GridSearchCV(
    rf, param_grid_rf, scoring='roc_auc',
    cv=3, n_jobs=-1, verbose=1
)

grid_rf.fit(X_train, y_train)

print("✅ Best RandomForest Params:", grid_rf.best_params_)
results.append(evaluate_model("Random Forest", grid_rf.best_estimator_, X_test, y_test))
```

```python
# ------------------------------------------
# 4️⃣ Compare All Models
# ------------------------------------------
results_df = pd.DataFrame(results).sort_values('ROC_AUC', ascending=False)
print("\n\n🏆 Model Comparison:\n")
print(results_df)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, ConfusionMatrixDisplay

# ==========================================
# 🧩 1. ROC Curve Comparison
# ==========================================

plt.figure(figsize=(8,6))
for name, model in [
    ("Logistic Regression", grid_lr.best_estimator_),
    ("Random Forest", grid_rf.best_estimator_),
    # ("XGBoost", grid_xgb.best_estimator_)
]:
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.3f})')

plt.plot([0,1], [0,1], 'k--')
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# ==========================================
# 🧩 2. Confusion Matrix for Each Model
# ==========================================
for name, model in [
    ("Logistic Regression", grid_lr.best_estimator_),
    ("Random Forest", grid_rf.best_estimator_),
    # ("XGBoost", grid_xgb.best_estimator_)
]:
    preds = model.predict(X_test)
    disp = ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
    disp.ax_.set_title(f"Confusion Matrix: {name}")
    plt.show()

# ==========================================
# 🧩 3. Bar Chart: Model Performance Comparison
# ==========================================
results_df.plot(
    x='Model',
    y=['Accuracy', 'ROC_AUC'],
    kind='bar',
    figsize=(8,5),
    title='Model Comparison (Accuracy & ROC-AUC)',
    rot=0
)
plt.grid(axis='y')
plt.show()

# ==========================================
# 🧩 4. Feature Importance (RandomForest & XGBoost)
# ==========================================
# Extract feature names after preprocessing
ohe = grid_rf.best_estimator_.named_steps['preprocess']
# Get feature names from the OneHotEncoder for categorical features
encoded_cat_cols = ohe.named_transformers_['trf'].get_feature_names_out(['batting_team', 'bowling_team', 'city'])
# Get the names of the numerical columns
numeric_cols = ['runs_left', 'balls_left', 'wickets', 'target_runs', 'crr', 'rrr']
# Combine all feature names
all_features = np.concatenate([encoded_cat_cols, numeric_cols])

# --- Random Forest Importance ---
rf_model = grid_rf.best_estimator_.named_steps['model']
importances = rf_model.feature_importances_
indices = np.argsort(importances)[-15:]  # Top 15 features

plt.figure(figsize=(8,5))
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), all_features[indices])
plt.title("Top Feature Importances (Random Forest)")
plt.xlabel("Importance")
plt.show()

# # --- XGBoost Importance ---
# xgb_model = grid_xgb.best_estimator_.named_steps['model']
# importances_xgb = xgb_model.feature_importances_
# indices_xgb = np.argsort(importances_xgb)[-15:]

# plt.figure(figsize=(8,5))
# plt.barh(range(len(indices_xgb)), importances_xgb[indices_xgb], align='center', color='orange')
# plt.yticks(range(len(indices_xgb)), all_features[indices_xgb])
# plt.title("Top Feature Importances (XGBoost)")
# plt.xlabel("Importance")
# plt.show()
```
